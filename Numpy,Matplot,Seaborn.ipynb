```python
##using matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(1,10,5)
y=np.sin(x)
plt.plot(x,y,color='black',linestyle='dotted',label=np.sin(x))
plt.axis([0,10,-1,1])
plt.title("My first graph")
plt.legend(loc='lower left')
ax=plt.axes()
ax.set(xlabel="x",ylabel="y")
```

```python
##errorbar
z=np.linspace(0,10,50)
dy=0.8
p=np.sin(z)+dy*np.random.randn(50)
plt.errorbar(z,p,yerr=dy,fmt=".k")
```

```python
plt.style.available[0:]
```

```python
fig=plt.figure(figsize=(12,8))
for i in range(6):
    fig.add_subplot(3,2,i+1)
    plt.style.use(plt.style.available[i])
    plt.plot(z,p)
    plt.text(s=plt.style.available[i],x=5,y=1,color='red')
```
