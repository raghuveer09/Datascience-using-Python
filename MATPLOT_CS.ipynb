```python
#axis grids
tit=sns.load_dataset("titanic")
```

```python
g=sns.FacetGrid(tit,col="survived",row='sex')
g.despine(left=False)
g.set_ylabels("Survived")
g.set_xticklabels(rotation=45)
g.set_axis_labels("Survived","sex")
```
