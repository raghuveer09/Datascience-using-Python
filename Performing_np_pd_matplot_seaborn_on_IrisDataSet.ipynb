```python
##Performing all the actions on Iris Data Set using numpy,pandas,matplot,seaborn
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#loading the iris data set using seaborn and knowing the shape and dimensions of the data set
df=sns.load_dataset('iris')
df.head()
type(df)
df.shape
df.ndim
df.tail()
type(df)
```

```python
#performing visualizations on the iris dataset using matplot and seaborn
#scatter plot
plt.scatter('sepal_length','petal_length',c='g',data=df)
plt.xlabel("Xaxis sepal_length")
plt.ylabel("Yaxis petal_length")
plt.title("Sepal vs Petal_Length")
```

```python
# to savethe above plot 
plt.savefig('SEPAL VS PETAL LENGTH.png')
```

```python
#scatter plot
plt.scatter('sepal_width','petal_width',c='g',data=df)
plt.xlabel("Xaxis sepal_width")
plt.ylabel("Yaxis petal_width")
plt.title("Sepal vs Petal_width")
```

```python
#creating subplots 
plt.subplot(2,2,1)
plt.scatter('sepal_length','sepal_width',c='r',data=df)
plt.subplot(2,2,2)
plt.scatter('petal_length','petal_width',c='g',data=df)
plt.subplot(2,2,3)
plt.scatter('petal_length','sepal_length',c='b',data=df)
plt.subplot(2,2,4)
plt.scatter('petal_width',"sepal_width",c='black',data=df)
```

```python
#barplot
plt.bar("species","petal_length",color='g',data=df)
plt.xlabel("species")
plt.ylabel("petal_length")
plt.title("species  petal_length")
```

```python
#barplot
plt.bar("species","petal_width",color='b',data=df)
plt.xlabel("species")
plt.ylabel("petal_width")
plt.title("species  petal_width")
```

```python
#subplots of barplot
plt.subplot(2,2,1)
plt.bar("species","petal_length",color='r',data=df)
plt.xlabel("species")
plt.ylabel("petal_length")
plt.subplot(2,2,2)
plt.bar("species","petal_width",color='b',data=df)
plt.xlabel("species")
plt.ylabel("petal_width")
plt.subplot(2,2,3)
plt.bar("species","sepal_length",color='black',data=df)
plt.xlabel("species")
plt.ylabel("sepal_length")
plt.subplot(2,2,4)
#barplot
plt.bar("species","sepal_width",color='g',data=df)
plt.xlabel("species")
plt.ylabel("sepal_width")
```
